```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Cargar el archivo
data = np.load('sph_grid_opt.npz')
positions = data['positions']

# Obtener informacin bsica
num_frames = positions.shape[0]
num_particles = positions.shape[1]

# Seleccionar 4 frames distribuidos uniformemente (Inicio, 33%, 66%, Final)
indices_to_plot = [0, num_frames // 3, (2 * num_frames) // 3, num_frames - 1]

# Configurar la figura (2x2 subplots)
fig = plt.figure(figsize=(16, 12))

for i, frame_idx in enumerate(indices_to_plot):
    ax = fig.add_subplot(2, 2, i + 1, projection='3d')

    # Extraer coordenadas del frame actual
    pos = positions[frame_idx]
    x = pos[:, 0]
    y = pos[:, 2]
    z = pos[:, 1]

    # Graficar
    # Usamos c=y para colorear por altura, s=0.1 para que las partculas se vean finas
    scatter = ax.scatter(x, y, z, c=y, cmap='viridis', s=0.2, alpha=0.6)

    # Esttica
    ax.set_title(f'Frame {frame_idx}', fontsize=14)
    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_zlabel('Z')

    # Fijar lmites para que la "caja" no cambie de tamao entre frames
    # Basado en el script anterior, la caja era de tamao 4.0
    ax.set_xlim(0, 4)
    ax.set_ylim(0, 4)
    ax.set_zlim(0, 4)

    # Ajustar vista
    ax.view_init(elev=20, azim=45)

plt.tight_layout()

# Guardar con alta resolucin
filename = 'sph_simulation_frames.png'
plt.savefig(filename, dpi=400)

print(f"Imagen guardada: {filename}")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Cargar el archivo
data = np.load('sph_grid_opt.npz')
positions = data['positions']

# Obtener informacin bsica
num_frames = positions.shape[0]
num_particles = positions.shape[1]

# Seleccionar 4 frames distribuidos uniformemente (Inicio, 33%, 66%, Final)
indices_to_plot = [0, (80+num_frames) // 3, (1 * num_frames) // 2, num_frames - 1]

# Configurar la figura (2x2 subplots)
fig = plt.figure(figsize=(16, 12))

for i, frame_idx in enumerate(indices_to_plot):
    ax = fig.add_subplot(2, 2, i + 1, projection='3d')

    # Extraer coordenadas del frame actual
    pos = positions[frame_idx]
    x = pos[:, 0]
    y = pos[:, 2]
    z = pos[:, 1]

    # Graficar
    # Usamos c=y para colorear por altura, s=0.1 para que las partculas se vean finas
    scatter = ax.scatter(x, y, z, color='c', s=0.2, alpha=0.6)

    # Esttica
    ax.set_title(f'Frame {frame_idx}', fontsize=14)
    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_zlabel('Z')

    # Fijar lmites para que la "caja" no cambie de tamao entre frames
    # Basado en el script anterior, la caja era de tamao 4.0
    ax.set_xlim(0, 4)
    ax.set_ylim(0, 4)
    ax.set_zlim(0, 4)

    # Ajustar vista
    ax.view_init(elev=20, azim=45)

plt.tight_layout()

# Guardar con alta resolucin
filename = 'sph_simulation_frames.png'
plt.savefig(filename, dpi=400)

print(f"Imagen guardada: {filename}")
```
